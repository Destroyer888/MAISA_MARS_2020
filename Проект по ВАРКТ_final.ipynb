```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
import pandas as pd
```

```python
df = pd.read_csv('Data_new.csv', sep=';')
Altitude = df['AltitudeFromTerrain']
Time = df['Time']
Velocity = df['Velocity']
Acceleration = df['Acceleration']
Mass = df['Mass']

plt.plot(Time, Velocity)
plt.ylabel(r'$V$')
plt.xlabel(r'$t$')
plt.title(r'    (KSP)')
plt.grid(True)
plt.show()

plt.plot(Time, Acceleration)
plt.ylabel(r'$a$')
plt.xlabel(r'$t$')
plt.title(r'  (KSP)  ')
plt.grid(True)
plt.show()

plt.plot(Altitude)
plt.ylabel(r'$H$')
plt.xlabel(r'$t$')
plt.title(r'    (KSP)')
plt.grid(True)
plt.show()

plt.plot(Time, Mass*2500)
plt.ylabel(r'$M$')
plt.xlabel(r'$t$')
plt.title(r'    (KSP)')
plt.grid(True)
plt.show()
```

```python
G = 6.6743*10**(-11)
Mz = 5.974*10**24
m0 = 571585
n1 = 1900
n2 = 20
u = 3000
fuel1 = 500265
fuel2 = 20830
msuh1 = 38247
msuh2 = 2243
m1 = msuh1 + fuel1
m2 = msuh2 + fuel2
Mgraph = []
V = []
Tacc = np.arange(1, 400, 1)
v0 = 0
F1 = 4248400
F2 = 749200

def Ftyag(m, AltitudeFromTerrain):
  return G*((Mz*m) / (6400000+AltitudeFromTerrain)**2)

for t in range(1, 254):
    m = m0 - n1 * t
    Mgraph.append(m)
    v1 = (v0 - u * np.log(m / m0) - (G * Mz * np.log(m / m0)) / ((6400000) ** 2))
    d = v1 / 1000 * 4
    V.append(d)
    v0 = v1
    
m0 = m0 - fuel1 - msuh1

for t in range(254, 1095):
    m = m0 - n2 * t
    Mgraph.append(m)
    v1 = (v0 - u * np.log(m / m0) - (G * Mz * np.log(m / m0)) / ((6400000) ** 2))
    d = v1 / 1000 * 4
    V.append(d)
    v0 = v1
    
m0 = m0 - fuel2 - msuh2


Acc = [0]*424
T = np.arange(1, 425, 1)
Th = np.arange(1,501,1)
H = [0]*500

for t in range(424):
    if (t < 254):
        Acc[t] = ((u*n1*t - Ftyag(m1-(4*1120+1818)*(t+1), Altitude[t])) / m1) / 80
        H[t+1] = (Acc[t]*((t+1)**2))/200 
    if (t >= 254):
        Acc[t] = ((u*n2*t - Ftyag(m2-2281*2*(t+1), Altitude[t] )) / m2) / 80
        H[t+1] = (Acc[t]*((t+1)**2))/200

plt.plot(T, Acc)
plt.title("", fontsize = 17)
plt.ylabel(" (/^2)")
plt.xlabel(" (c)")
plt.grid(True)
plt.show()

plt.plot(Th, H)
plt.title("  ", fontsize = 17)
plt.ylabel(" ()")
plt.xlabel(" (c)")
plt.grid(True)
plt.show()

plt.plot(Mgraph)
plt.ylabel(r'$M()$')
plt.xlabel(r'$t$')
plt.title(r'    ')
plt.grid(True)
plt.show()

plt.plot(V)
plt.ylabel(r'$V(/c)$')
plt.xlabel(r'$t$')
plt.title(r'    ')
plt.grid(True)
plt.show()
```
